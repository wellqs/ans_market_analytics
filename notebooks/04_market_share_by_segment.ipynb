```python
import sqlite3
from pathlib import Path

DB_PATH = Path("..") / "database" / "ans.db"

conn = sqlite3.connect(DB_PATH)
cursor = conn.cursor()

print("Banco:", DB_PATH.resolve())
```

```python
query = """
WITH operadora_segmento AS (
    SELECT
        competencia,
        registro_ans,
        modalidade_grupo,
        MAX(beneficiarios) AS beneficiarios
    FROM si_beneficiarios_operadora
    GROUP BY
        competencia,
        registro_ans,
        modalidade_grupo
),

market_share AS (
    SELECT
        competencia,
        registro_ans,
        modalidade_grupo,
        beneficiarios,

        ROUND(
            beneficiarios * 100.0 /
            SUM(beneficiarios) OVER (
                PARTITION BY competencia, modalidade_grupo
            ),
            4
        ) AS market_share_pct,

        RANK() OVER (
            PARTITION BY competencia, modalidade_grupo
            ORDER BY beneficiarios DESC
        ) AS ranking
    FROM operadora_segmento
)

SELECT *
FROM market_share
WHERE ranking <= 10
ORDER BY competencia, modalidade_grupo, ranking;
"""

cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```
