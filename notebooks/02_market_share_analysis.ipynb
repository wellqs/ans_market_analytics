```python
query = """
WITH operadora_mes AS (
    SELECT
        b.competencia,
        b.registro_ans,
        o.razao_social,
        MAX(b.beneficiarios) AS beneficiarios
    FROM si_beneficiarios_operadora b
    LEFT JOIN si_operadora o
        ON o.registro_ans = b.registro_ans
    GROUP BY
        b.competencia,
        b.registro_ans,
        o.razao_social
)

SELECT
    competencia,
    registro_ans,
    razao_social,
    beneficiarios,

    ROUND(
        beneficiarios * 100.0 /
        SUM(beneficiarios) OVER (PARTITION BY competencia),
        4
    ) AS market_share_pct,

    RANK() OVER (
        PARTITION BY competencia
        ORDER BY beneficiarios DESC
    ) AS ranking

FROM operadora_mes
ORDER BY ranking
LIMIT 20;
"""

cursor.execute(query)

rows = cursor.fetchall()

for row in rows:
    print(row)
```
