```python
import sqlite3
from pathlib import Path

DB_PATH = Path("..") / "database" / "ans.db"

conn = sqlite3.connect(DB_PATH)
cursor = conn.cursor()

print("Banco conectado:", DB_PATH.resolve())
```

```python
query = """
WITH operadora_mes AS (
    SELECT
        competencia,
        registro_ans,
        MAX(beneficiarios) AS beneficiarios
    FROM si_beneficiarios_operadora
    GROUP BY
        competencia,
        registro_ans
)

SELECT
    competencia,
    registro_ans,
    beneficiarios
FROM operadora_mes
ORDER BY competencia, beneficiarios DESC;
"""

cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```

```python
query = """
WITH operadora_mes AS (
    SELECT
        competencia,
        registro_ans,
        MAX(beneficiarios) AS beneficiarios
    FROM si_beneficiarios_operadora
    GROUP BY competencia, registro_ans
),

market_share AS (
    SELECT
        competencia,
        registro_ans,
        beneficiarios,

        ROUND(
            beneficiarios * 100.0 /
            SUM(beneficiarios) OVER (PARTITION BY competencia),
            4
        ) AS market_share_pct
    FROM operadora_mes
)

SELECT *
FROM market_share
ORDER BY competencia, market_share_pct DESC;
"""

cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```

```python
query = """
WITH operadora_mes AS (
    SELECT
        competencia,
        registro_ans,
        MAX(beneficiarios) AS beneficiarios
    FROM si_beneficiarios_operadora
    GROUP BY competencia, registro_ans
)

SELECT
    competencia,
    registro_ans,
    beneficiarios,

    RANK() OVER (
        PARTITION BY competencia
        ORDER BY beneficiarios DESC
    ) AS ranking
FROM operadora_mes
ORDER BY competencia, ranking;
"""

cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```

```python
query = """
WITH operadora_mes AS (
    SELECT
        competencia,
        registro_ans,
        MAX(beneficiarios) AS beneficiarios
    FROM si_beneficiarios_operadora
    GROUP BY competencia, registro_ans
),

growth AS (
    SELECT
        competencia,
        registro_ans,
        beneficiarios,

        LAG(beneficiarios) OVER (
            PARTITION BY registro_ans
            ORDER BY competencia
        ) AS beneficiarios_mes_anterior
    FROM operadora_mes
)

SELECT
    competencia,
    registro_ans,
    beneficiarios,
    beneficiarios_mes_anterior,

    beneficiarios - beneficiarios_mes_anterior AS crescimento_absoluto
FROM growth
ORDER BY competencia, crescimento_absoluto DESC;
"""

cursor.execute(query)

for row in cursor.fetchall():
    print(row)
```

## Dinâmica de Mercado

A análise de dinâmica permite identificar mudanças estruturais no mercado de saúde suplementar.

Indicadores analisados:

- crescimento de beneficiários por operadora
- evolução do market share
- mudança de ranking entre operadoras

Essas métricas ajudam a identificar operadoras que estão expandindo participação de mercado e aquelas que estão perdendo espaço competitivo.
